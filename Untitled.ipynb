```python
df = pd.read_csv("/Users/glmarschmann/Data/BioFuels/Munson-McGee-SAG-Statistics_Class.csv")
```

```python
df = df.dropna(subset=['Predicted_respiration_rate_fmol_O2_per_hr'])
```

```python
df["resp"] = df["Predicted_respiration_rate_fmol_O2_per_hr"]/(60**2)*100
```

```python
ax = plt.gca()
#ax = sns.stripplot(x="class", y="resp", palette =["k", 'k', 'k'], data=df, dodge=False, size=1, label='')
ax = sns.boxplot(x="class", y="resp", data=df, palette =["w", 'w', '#3182bd'], dodge=False, showfliers=True, order=["Alphaproteobacteria", "Gammaproteobacteria", "Bacteroidia"])
ax.set(yscale="log")
plt.ylabel("Per-cell respiration rate (1e-17 CO2/cell/s)")
plt.xlabel("")
plt.savefig("/Users/glmarschmann/Data/BioFuels/single_cell_respiration.png")
```
